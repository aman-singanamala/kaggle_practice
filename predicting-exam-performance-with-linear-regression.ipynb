## Just consider math score, reading score, writing score in a new Dataframe variable X


```python
X=data[['math score','reading score', 'writing score']]
```

```python
X.head(n=5)
```

```python
Y_math= X['math score']
Y_reading= X['reading score']
Y_writing= X['writing score']
```

## Scale the datasets

```python
scaler= StandardScaler()
X= pd.DataFrame(scaler.fit_transform(X), columns= X.columns)
```

```python
X
```

```python
X_math= X.iloc[:, [1,2]]    # drop matn score
X_reading= X.iloc[:, [0,2]] # drop reading score
X_writing= X.iloc[:, [0,1]] # drop writing score
```

## Split the data into train and test sets

```python
X_math_train, X_math_test , Y_math_train, Y_math_test = train_test_split(X_math,
                                                                         Y_math, 
                                                                         train_size= 0.7)


X_reading_train, X_reading_test , Y_reading_train, Y_reading_test = train_test_split(X_reading, 
                                                                                     Y_reading, 
                                                                                     train_size= 0.7)


X_writing_train, X_writing_test , Y_writing_train, Y_writing_test = train_test_split(X_writing, 
                                                                                     Y_writing, 
                                                                                     train_size= 0.7)
```

```python
math_model= LinearRegression()
reading_model= LinearRegression()
writing_model= LinearRegression()
math_model.fit(X_math_train, Y_math_train)
reading_model.fit(X_reading_train, Y_reading_train)
writing_model.fit(X_writing_train, Y_writing_train)
```

```python
math_R2= math_model.score(X_math_test, Y_math_test)
reading_R2= reading_model.score(X_reading_test, Y_reading_test)
writing_R2= writing_model.score(X_writing_test, Y_writing_test)
```

```python
print(f" Math r^2 value : { math_R2}")
print(f" Reading r^2 value : { reading_R2}")
print(f" Writing r^2 value : { reading_R2}")
```

## Reading and writing are predicted somewhat easily than Math score
